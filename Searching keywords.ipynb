### Data Dictionary Term

```python
dict_term = pd.read_csv('./Preference_Term.csv')
dict_term.drop(columns = ["Gender", "Gender_BirthSex", "Unnamed: 13"], inplace = True)
dict_term['Special_Term'] = np.where(dict_term['Province'] == 'สตูล', 'ไฮ',
                             np.where(dict_term['Province'] == 'ระยอง', 'ไฮเงี่ยน',
                             np.where(dict_term['Province'] == 'ชุมพร', 'ยสตน', np.nan)))
```

```python
df_tweet = pd.read_csv("./data/df_tweet_for_searching.csv")
```

```python
for keywords_type in list(dict_term[['Special_Term']].columns):
    selected_search_term = dict_term.dropna(subset = [keywords_type])
    result_dict = {}
    for text_body, tweet_id in zip(df_tweet.text, df_tweet.id):
        finding_list = []
        key_list = list(selected_search_term[keywords_type])
        key_list.sort(key = len, reverse = True)
        print(key_list)
        
        for keyword in key_list:
            if str(keyword) in text_body:
                finding_list.append(keyword)
                text_body = text_body.replace(str(keyword), "")
                   
        result_dict[tweet_id] = finding_list
```

```python
for keywords_type in list(dict_term.columns):
    selected_search_term = dict_term.dropna(subset = [keywords_type])
    result_dict = {}

    for text_body, tweet_id in zip(df_tweet.text, df_tweet.id):
        finding_list = []
        key_list = list(selected_search_term[keywords_type])
        key_list.sort(key = len, reverse = True)
        
        for keyword in key_list:
            
            if str(keyword) in text_body:
                finding_list.append(keyword)
                text_body = text_body.replace(str(keyword), "")
                    
        result_dict[tweet_id] = finding_list

    search_result_df = pd.DataFrame(result_dict.items())
    search_result_df.rename(columns = {0: "id", 1 : keywords_type}, inplace = True)

    df_tweet = pd.merge(df_tweet, search_result_df, on = "id")
```
